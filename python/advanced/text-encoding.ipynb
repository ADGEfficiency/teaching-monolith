## Bits & bytes

Bit = [0, 1]

Byte = 8 bits

Byte = $2^{8} - 1 = 256$ integers (0 to 255)

## Hexadecimal

https://simple.wikipedia.org/wiki/Hexadecimal_numeral_system

Decimal = base 10
- symbols 0-9

Hexadecimal = base 16
- symbols 0-9 & A-F

Hexadecimal numbers are sometimes written with a "h" after or "0x" before the number
- 63h = 63 hexadecimal
- 0x64 = 64 hexadecimal




## Strings & characters

String = sequence of characters

Charater = A, a, ? etc

Characters need to be encoded into binary before a machine can understand them

## First there was ASCII

ASCII = American Standard Code for Information Interchange
- defines 128 characters using 128 integers
- means one byte has more than enough space to encode any ASCII character

```
0  NUL (null) 
9  TAB (horizontal tab) 
32  SPACE
48  0
57  9
65  A
97 a
127  DEL
```

See this [ASCII Table](https://www.cs.cmu.edu/~pattis/15-1XX/common/handouts/ascii.html) for the complete 128

## Then Unicode

The problem with ASCII is that it is English characters only

Unicode = a single character set that could represent every character in every imaginable language

Characters are represented as **code-points**:
- U+00639
- U = Unicode
- numbers are hexadecimal

Unicode = superset of ASCII
- ASCII characters are the same in Unicode

## Finally UTF-8

UTF-8 is dominant on the web, and since HTML5, it has been the recommended encoding

Unicode characters don't generally fit into one 8-bit byte
- there are numerous ways of storing Unicode characters in byte sequences

UTF-8 = system for storing and representing these code-points in one to four bytes
- 1 byte: Standard ASCII
- 2 bytes: Arabic, Hebrew, most European scripts (most notably excluding Georgian)
- 3 bytes: BMP
- 4 bytes: All Unicode characters

The '8-bit' encoding means that the individual bytes of the encoding use 8 bits
